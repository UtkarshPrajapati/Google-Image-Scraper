```python
import requests
from bs4 import BeautifulSoup
import os
```

```python
sdir="images/"
if not os.path.exists(sdir): os.makedirs(sdir)
```

```python
query=input() or "Utkarsh Prajapati"
resp=BeautifulSoup(requests.get(f"https://www.google.com/search?q={query}&tbm=isch").content
,'html.parser')
resp
```

```python
links=resp.find_all("img",{"class":"DS1iW"})
links
```

```python
imgs=[i["src"] for i in links]
imgs
```

```python
imgdicts={i:requests.get(i).content for i in imgs}
imgdicts
```

```python
for i in imgdicts:
    with open(os.path.join(sdir,f"{query}_{imgs.index(i)+1}.jpg"),"wb") as f:
        f.write(imgdicts[i])
```
